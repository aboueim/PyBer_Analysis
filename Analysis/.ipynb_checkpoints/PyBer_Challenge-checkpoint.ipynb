# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
rides_total = pyber_data_df.groupby('type').count()['ride_id']
rides_total
```

```python
# 2. Get the total drivers for each city type
drivers_total = city_data_df.groupby('type').sum()['driver_count']
drivers_total
```

```python
#  3. Get the total amount of fares for each city type
fares_total = pyber_data_df.groupby('type').sum()['fare']
fares_total
```

```python
#  4. Get the average fare per ride for each city type. 
fare_per_ride = fares_total / rides_total
fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
fare_per_driver = fares_total / drivers_total
fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({"Total Rides": rides_total, "Total Drivers": drivers_total, "Total Fares": fares_total, "Average Fare per Ride": fare_per_ride, "Average Fare per Driver": fare_per_driver})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df.loc[:,"Total Rides"] = pyber_summary_df["Total Rides"].map('{:,.0f}'.format)
pyber_summary_df.loc[:,"Total Drivers"] = pyber_summary_df["Total Drivers"].map('{:,.0f}'.format)
pyber_summary_df.loc[:,"Total Fares"] = '$' + pyber_summary_df["Total Fares"].map('{:,.2f}'.format)
pyber_summary_df.loc[:,"Average Fare per Ride"] = '$' + pyber_summary_df["Average Fare per Ride"].map('{:,.2f}'.format)
pyber_summary_df.loc[:,"Average Fare per Driver"] = '$' + pyber_summary_df["Average Fare per Driver"].map('{:,.2f}'.format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head(10)
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
city_type_sum_fares = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
city_type_sum_fares
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
city_type_sum_fares = city_type_sum_fares.reset_index()
city_type_sum_fares
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
city_type_sum_fares_pivot = city_type_sum_fares.pivot(index = "date", columns = 'type', values = 'fare')
city_type_sum_fares_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
city_type_sum_fares_pivot_date_range = city_type_sum_fares_pivot.loc["2019-01-01":"2019-04-28"]
city_type_sum_fares_pivot_date_range
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
city_type_sum_fares_pivot_date_range.index = pd.to_datetime(city_type_sum_fares_pivot_date_range.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
city_type_sum_fares_pivot_date_range.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
city_type_sum_fares_pivot_date_range_resample = city_type_sum_fares_pivot_date_range.resample("W").sum()
city_type_sum_fares_pivot_date_range_resample.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
import matplotlib.dates as mdates
from matplotlib.ticker import MultipleLocator
import datetime

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

urban_sum_fares = city_type_sum_fares_pivot_date_range_resample['Rural']
suburban_sum_fares = city_type_sum_fares_pivot_date_range_resample['Suburban']
rural_sum_fares = city_type_sum_fares_pivot_date_range_resample['Urban']

fig, ax = plt.subplots(figsize = (25, 8))

ax.plot (city_type_sum_fares_pivot_date_range_resample.index, urban_sum_fares, label = 'rural')
ax.plot (city_type_sum_fares_pivot_date_range_resample.index, suburban_sum_fares, label = 'suburban')
ax.plot (city_type_sum_fares_pivot_date_range_resample.index, rural_sum_fares, label = 'urban')

ax.set_xlim([datetime.date(2019,1,6), datetime.date(2019,4,28)])
ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(ax.xaxis.get_major_locator()))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.set_ylabel('Fare ($USD)')
ax.set_title('Total Fare by City Type')
ax.legend(title = 'type', loc = 'center')
plt.savefig('PyBer_fare_summary.png')
```
